```python
weather=pd.read_csv('data/weather.csv')
weather.head()
```

```python
weather_melt=pd.melt(weather,id_vars=['id','year','month','element'],var_name='days',value_name='temp')
weather_melt.head()
```

```python
weather_pivot=weather_melt.pivot_table(index=['id','year','month','days'],columns='elements',values='temp')
weather_pivot.head()
```
