# vectorizng

method of different log

```python
x=np.arange(1,101)
y=np.log(x)
plt.plot(x,y)
plt.show()
```

```python

import math
x=np.arange(1,101)
#y=list(map(lambda a:math.log(a),x))
y=list(map(math.log,x))
plt.plot(x,y)
plt.show()
```

```python

import math
x=np.arange(1,101)
y=[math.log(i) for i in x ]
plt.plot(x,y)
plt.show()
```

```python
#vectorize function
import math
x=np.arange(1,101)
new_log=np.vectorize(math.log)
y=new_log(x)
plt.plot(x,y)
plt.show()
```

```python

import math
def sigmoid(z):
    return 1/(1+math.exp(-z))
x=np.arange(-10,10,0.1)
y=[sigmoid(i) for i in x ]
plt.plot(x,y)
plt.show()
```
